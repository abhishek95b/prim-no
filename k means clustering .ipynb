```python
# import k means function
from sklearn.cluster import KMeans
```

```python
# assign no of cluster for group
k=KMeans(n_clusters=4)
```

```python
k
```

```python
#  fit data for pred
y_pred=k.fit_predict(a[['Age','Income($)']])
```

```python
y_pred
```

```python
# ctreat new column clustrer in dataset for grouping according income
a['cluster']=y_pred
```

```python
a
```

```python
# cenrers for grouping
k.cluster_centers_
```

```python
# creat cluster and grouping data according to income and indicate by diff colour

k1=a[a.cluster==0]
k2=a[a.cluster==1]
k3=a[a.cluster==2]
k4=a[a.cluster==3]
plt.scatter(k1.Age,k1['Income($)'],color='green',marker='*',label='income')

plt.scatter(k2.Age,k2['Income($)'],color='red',marker='*',label='income')

plt.scatter(k3.Age,k3['Income($)'],color='black',marker='*',label='income')

plt.scatter(k4.Age,k4['Income($)'],color='blue',marker='*',label='income')

plt.scatter(k.cluster_centers_[:,0],k.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Age')
plt.ylabel('income')
```
